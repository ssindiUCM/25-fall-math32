## ðŸŽ² Betting on Dice. (From Last Time)

**Goal:** Simulate rolling two dice and compute your winnings or losses.

You win \$1000 if the sum of the dice is 11 or 12.  

Otherwise, you lose \$100.

```python
# ðŸŽ² Dice Simulation Exercise
# Goal: Simulate rolling two dice and compute winnings/losses

import random
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# Step 1: Define parameters
# ----------------------------
n_trials = 1000000  # number of simulations
winnings = []      # list to store results

# ----------------------------
# Step 2: Run the simulation
# ----------------------------
for i in range(n_trials):
    # Roll two dice
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    
    # Compute the sum
    total = die1 + die2
    
    # Compute the outcome
    if total >= 11:
        winnings.append(1000)
    else:
        winnings.append(-100)

# ----------------------------
# Step 3: Analyze the results
# ----------------------------
winnings = np.array(winnings)

# Empirical probabilities
unique, counts = np.unique(winnings, return_counts=True)
pmf_empirical = counts / n_trials

# Theoretical PMF
pmf_theoretical = {
    1000: 3/36,   # (11 or 12) â†’ 2 + 1 = 3 outcomes
    -100: 33/36
}

# ----------------------------
# Step 4: Expected Value (E[Y])
# ----------------------------
# Theoretical expected value
E_Y_theoretical = sum(val * p for val, p in pmf_theoretical.items())

# Empirical expected value (sample mean)
E_Y_empirical = winnings.mean()

print("Expected Value of Winnings:")
print(f"  Theoretical E[Y] = {E_Y_theoretical:.2f}")
print(f"  Empirical   E[Y] â‰ˆ {E_Y_empirical:.2f}\n")

# ----------------------------
# Step 5: Display results
# ----------------------------
print("Empirical PMF:")
for val, prob in zip(unique, pmf_empirical):
    print(f"  P(Y = {val}) â‰ˆ {prob:.4f}")

print("\nTheoretical PMF:")
for val, prob in pmf_theoretical.items():
    print(f"  P(Y = {val}) = {prob:.4f}")

# ----------------------------
# Step 6: Visualize (clearer style)
# ----------------------------
fig, ax = plt.subplots(figsize=(6,4))

# Bars for empirical PMF
ax.bar(unique.astype(str), pmf_empirical, alpha=0.6, color='skyblue', label='Empirical (Simulation)')

# Dots + dashed line for theoretical PMF
x_theo = [str(k) for k in pmf_theoretical.keys()]
y_theo = list(pmf_theoretical.values())
ax.plot(x_theo, y_theo, 'o--', color='red', label='Theoretical (Exact)', markersize=10, lw=2)

ax.set_title("PMF of Winnings")
ax.set_xlabel("Winnings ($)")
ax.set_ylabel("Probability")
ax.legend()
ax.grid(alpha=0.2)
plt.show()
```

## ðŸŸ¦ Squaring a Uniform Distribution

Let $X \sim \text{Uniform}(0, 1)$.

We define a new random variable:
$Y = X^2$

### Goal:
1. Generate many samples of $X$ using simulation.  
2. Transform each sample to get $Y = X^2$.  
3. Plot the empirical PDF and CDF of $Y$, and compare them with the theoretical curves.

---

### Theoretical Result:

Start from the definition of the CDF:

$$
F_Y(y) = P(Y \le y) = P(X^2 \le y) = P(-\sqrt{y} \le X \le \sqrt{y})
$$

> **Note:** Since $X \sim \text{Uniform}(0,1)$, only the positive side matters. The lower limit $-\sqrt{y}$ is outside the support of $X$. So we simplify:

$$
F_Y(y) = P(X \le \sqrt{y}) = F_X(\sqrt{y}) = \sqrt{y}, \quad 0 \le y \le 1
$$

Differentiating with respect to $y$ gives the PDF: 
$$f_Y(y) = \frac{d}{dy} F_Y(y) = \frac{d}{dy} F_X(\sqrt{y}) = \frac{1}{2\sqrt{y}} f_x(\sqrt{y}).$$

Because the PDF of a Uniform(0,1) random variable is:
$$
f_X(x) =
\begin{cases}
1, & 0 \le x \le 1 \\
0, & \text{otherwise,}
\end{cases}
$$

we have:
$$f_Y(y) =
\begin{cases}
\dfrac{1}{2\sqrt{y}}, & 0 \leq y \leq 1 \\
0, & \text{otherwise.}
\end{cases}
$$

**Observation:**  
- Squaring \( X \) "compresses" values near 1 and "stretches" values near 0, so the PDF is higher near 0.  
- The CDF rises quickly at small \( y \) and flattens near 1, reflecting this concentration of probability at lower values.


```python
# ---------------------------------
# Squaring a Uniform Random Variable with Empirical CDF Bars
# ---------------------------------

import random
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Simulation parameters
n_samples = 100000
X_values = []
Y_values = []

# Step 2: Generate uniform samples and compute Y = X^2
for i in range(n_samples):
    x = random.random()
    y = x**2
    X_values.append(x)
    Y_values.append(y)

# Step 3: Compute histogram for PDF
fig, axs = plt.subplots(2, 1, figsize=(7, 8))

counts, bins, patches = axs[0].hist(
    Y_values, bins=50, density=True, alpha=0.6, label="Empirical PDF"
)

# Step 4: Use first bin midpoint as safe lower limit for theoretical grid
bin_start = bins[0]
bin_end = bins[1]
y_start = 0.5 * (bin_start + bin_end)

# Step 5: Theoretical PDF and CDF
y_grid = np.linspace(y_start, 1, 400)
f_y = 1 / (2 * np.sqrt(y_grid))            # PDF of Y
F_y = np.sqrt(y_grid)                      # CDF of Y

# Step 6: Plot theoretical PDF
axs[0].plot(y_grid, f_y, 'r--', linewidth=2, label="Theoretical PDF")
axs[0].set_ylim(0, 1.2 * max(counts))
axs[0].set_title("PDF of Y = XÂ² where X ~ Uniform(0,1)")
axs[0].set_xlabel("y")
axs[0].set_ylabel("Density")
axs[0].legend()

# Step 7: Empirical CDF as bars
Y_sorted = np.sort(Y_values)
empirical_cdf = np.arange(1, n_samples + 1) / n_samples

# Convert sorted Y to bins for bar plotting
bin_width = 0.01  # choose a reasonable width
bins_cdf = np.arange(0, 1+bin_width, bin_width)
cdf_counts, _ = np.histogram(Y_sorted, bins=bins_cdf)
cdf_cumulative = np.cumsum(cdf_counts) / n_samples

axs[1].bar(bins_cdf[:-1], cdf_cumulative, width=bin_width, alpha=0.6, label="Empirical CDF")
axs[1].plot(y_grid, F_y, 'r--', linewidth=2, label="Theoretical CDF")
axs[1].set_title("CDF of Y = XÂ² where X ~ Uniform(0,1)")
axs[1].set_xlabel("y")
axs[1].set_ylabel("Cumulative Probability")
axs[1].legend()

plt.tight_layout()
plt.show()
```

## ðŸŽ² ðŸŽ² ðŸŽ² ðŸŽ² Maximum of 4 Dice

We roll four fair six-sided dice:

$$
X_1, X_2, X_3, X_4 \sim \text{Uniform}\{1,2,3,4,5,6\}
$$

Define a new random variable:

$$
Y = \max\{X_1, X_2, X_3, X_4\}
$$

### Goal:
1. Derive the theoretical PMF of \(Y\).  
2. Simulate many rolls and compute the empirical PMF.  
3. Compare the empirical PMF with the theoretical PMF.

---

### Theoretical PMF:

The CDF of the maximum of independent dice is:

$$
F_Y(k) = P(Y \le k) = P(X_1 \le k, X_2 \le k, X_3 \le k, X_4 \le k) = \left(\frac{k}{6}\right)^4
$$

The PMF is the difference of CDF values:

$$
P(Y = k) = F_Y(k) - F_Y(k-1) = \left(\frac{k}{6}\right)^4 - \left(\frac{k-1}{6}\right)^4, \quad k = 1,2,\dots,6
$$

**Observation:**  
- The PMF is skewed toward higher numbers because taking the maximum favors large outcomes.  
- Students can simulate this and plot the **empirical PMF as bars** with the **theoretical PMF as red dots** for comparison.


```python
import random
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# Step 1: Parameters
# ----------------------------
n_trials = 100000
max_values = []

# ----------------------------
# Step 2: Simulate rolls
# ----------------------------
for _ in range(n_trials):
    X1 = random.randint(1,6)
    X2 = random.randint(1,6)
    X3 = random.randint(1,6)
    X4 = random.randint(1,6)
    Y = max(X1, X2, X3, X4)
    max_values.append(Y)

# ----------------------------
# Step 3: Empirical PMF
# ----------------------------
unique, counts = np.unique(max_values, return_counts=True)
pmf_empirical = counts / n_trials

# ----------------------------
# Step 4: Theoretical PMF
# ----------------------------
pmf_theoretical = {k: (k/6)**4 - ((k-1)/6)**4 for k in range(1,7)}

# ----------------------------
# Step 5: Display results
# ----------------------------
print("Empirical PMF:")
for val, prob in zip(unique, pmf_empirical):
    print(f"  P(Y={val}) â‰ˆ {prob:.4f}")

print("\nTheoretical PMF:")
for val, prob in pmf_theoretical.items():
    print(f"  P(Y={val}) = {prob:.4f}")

# ----------------------------
# Step 6: Plot
# ----------------------------
plt.figure(figsize=(7,4))

# Empirical PMF as bars
plt.bar(unique, pmf_empirical, alpha=0.6, label="Empirical")

# Theoretical PMF as red dots
plt.plot(list(pmf_theoretical.keys()), list(pmf_theoretical.values()),
         'ro', markersize=8, label="Theoretical")

plt.title("PMF of the Maximum of Four Dice")
plt.xlabel("Maximum Value Y")
plt.ylabel("Probability")
plt.xticks(range(1,7))
plt.legend()
plt.show()
```

## ðŸŽ² Two Dice: Sum and Maximum

We are going to study the **joint distribution** of two dice using simulation.  

Let:  

- \(X_1\) = Number on the first die (Yellow)  
- \(X_2\) = Number on the second die (Blue)  

Define two new random variables:  

- **Sum:** $S = X_1 + X_2$
- **Maximum:** $M = \max(X_1, X_2)$

### Goal:

1. Simulate **many rolls** of the two dice.  
2. Compute the **Sum** \(S\) and **Maximum** \(M\) for each roll.  
3. Calculate the **empirical PMF** $P(S=s, M=m)$
4. Calculate the **empirical CDF** $P(S \leq s, M \leq m)$.  
5. Visualize both PMF and CDF as **3D bar plots**.  


```python
%matplotlib inline
import random
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ---------------------------------------------
# Simulation of Two Dice: Sum and Maximum
# ---------------------------------------------
n_trials = 100000
results = []

# Generate samples
for _ in range(n_trials):
    X1 = random.randint(1,6)
    X2 = random.randint(1,6)
    Y_sum = X1 + X2
    Y_max = max(X1, X2)
    results.append((Y_sum, Y_max))

results = np.array(results)
Y1_vals = results[:,0]  # Sum
Y2_vals = results[:,1]  # Max

s_vals = np.arange(2,13)   # Sum axis
m_vals = np.arange(1,7)    # Max axis

# ---------------------------------------------
# Empirical PMF
# ---------------------------------------------
pmf = np.zeros((len(m_vals), len(s_vals)))
for i, m in enumerate(m_vals):
    for j, s in enumerate(s_vals):
        pmf[i,j] = np.sum((Y1_vals==s) & (Y2_vals==m)) / n_trials

# ---------------------------------------------
# Empirical CDF
# ---------------------------------------------
cdf = np.zeros_like(pmf)
for i in range(len(m_vals)):
    for j in range(len(s_vals)):
        cdf[i,j] = np.sum(pmf[:i+1, :j+1])

# ---------------------------------------------
# 3D Bar Plots (Static)
# ---------------------------------------------
fig = plt.figure(figsize=(12,5))

# Meshgrid for axes
x_pos, y_pos = np.meshgrid(s_vals, m_vals)
x_pos = x_pos.ravel()    # Sum
y_pos = y_pos.ravel()    # Max
z_pos = np.zeros_like(x_pos)
dx = dy = 0.5

# ----- PMF -----
ax1 = fig.add_subplot(121, projection='3d')
dz = pmf.ravel()
ax1.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, shade=True, color='skyblue')
ax1.set_xlabel('Sum of Dice (S)')
ax1.set_ylabel('Maximum of Dice (M)')
ax1.set_zlabel('Probability')
ax1.set_title('Empirical PMF')
ax1.set_xticks(s_vals)
ax1.set_yticks(m_vals)
ax1.view_init(elev=30, azim=45)  # corner S=2, M=1 closest

# ----- CDF -----
ax2 = fig.add_subplot(122, projection='3d')
dz_cdf = cdf.ravel()
ax2.bar3d(x_pos, y_pos, z_pos, dx, dy, dz_cdf, shade=True, color='orange')
ax2.set_xlabel('Sum of Dice (S)')
ax2.set_ylabel('Maximum of Dice (M)')
ax2.set_zlabel('CDF')
ax2.set_title('Empirical CDF')
ax2.set_xticks(s_vals)
ax2.set_yticks(m_vals)
ax2.view_init(elev=30, azim=45)

plt.tight_layout()
plt.show()
```
