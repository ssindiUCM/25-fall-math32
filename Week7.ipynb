This goal of this notebook is to give you some example code with which to work with on calculating tail probabilities for normal distributions.

There are 3 examples here:
- Left Tail
- Right Tail
- Both Tails

Remember, when working with normal distributions there are three steps:

* Step 1: Is this a standard normal? N(0,1)
* Step 2: If no, then standardize by subtracting $\mu$ and dividing by $\sigma$.
* Step 3: Calculate the appropriate tail probability `norm.cdf()` to calculate the cumulative probability.


## Example 1 â€” Pressure Sensor

You are an engineer monitoring a pressure sensor. During normal operation, the pressure  
\(X\) follows:

$$
X \sim N(\mu=10, \sigma^2 = 0.25)
$$

**Question**: What is the probability that the pressure drops below 9?

---

**Step 1:** This is **not a standard normal**.

**Step 2:** Convert \(X\) to a standard normal \(Z\):

$$
P(X \leq 9) = P\left( \frac{X - \mu}{\sigma} \leq \frac{9-\mu}{\sigma} \right) = P(Z \leq \frac{9-\mu}{\sigma})
$$

Note that because $(\mu = 10)$ and $(\sigma = 0.5)$, we have \(Z = (9-10)/0.5 = -2\).

**Step 3:** Calculate the probability using Python:

`norm.cdf(-2)`

Below we carry out the calculation AND plot the PDF shaded in.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Parameters ---
mu = 10.0          # mean pressure
sigma = 0.5        # standard deviation
threshold = 9.0    # lower limit

# --- Step 1: Standardize ---
Z = (threshold - mu) / sigma
print(f"Standardized Z = {Z:.2f}")

# --- Step 2: Left-tail probability ---
P_left = norm.cdf(Z)
P_right = 1 - P_left

print(f"P(X <= {threshold}) = P(Z <= {Z:.2f}) = {P_left:.4f}")
print(f"P(X > {threshold}) = 1 - P(X <= {threshold}) = 1 - {P_left:.4f} = {P_right:.4f}")

# --- Step 3: Plot ---
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
pdf = norm.pdf(x, loc=mu, scale=sigma)

plt.figure(figsize=(8,5))

# PDF line
plt.plot(x, pdf, color='navy', lw=2, label=f'N({mu}, {sigma**2})')

# --- Shade succeed region (green) for X > threshold ---
x_fill_pass = np.linspace(threshold, mu + 4*sigma, 200)
plt.fill_between(x_fill_pass, 0, norm.pdf(x_fill_pass, mu, sigma),
                 color='green', alpha=0.5, label=f'Succeed (X > {threshold})')

# --- Shade fail region (red) for X <= threshold ---
x_fill_fail = np.linspace(mu - 4*sigma, threshold, 200)
plt.fill_between(x_fill_fail, 0, norm.pdf(x_fill_fail, mu, sigma),
                 color='red', alpha=0.5, label=f'Fail (X â‰¤ {threshold})')

# Threshold line
plt.axvline(threshold, color='black', linestyle='--', lw=1.5)

# Labels, title, legend
plt.title('Left-Tail Probability: Pressure Below Threshold')
plt.xlabel('Pressure')
plt.ylabel('Density f(x)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Example 2 â€” Manufacturing Tolerance

A machine produces rods whose diameters \(X\) follow:

$$
X \sim N(\mu = 5.0 \text{ mm}, \sigma^2 = 0.04)
$$

**Question:** What is the probability that a randomly chosen rod **exceeds 5.1 mm** in diameter?

---

**Step 1:** Standardize \(X\) to a standard normal \(Z\):

$$
Z = \frac{X - \mu}{\sigma} = \frac{5.1 - 5.0}{\sqrt{0.04}} = \frac{0.1}{0.2} = 0.5
$$

---

**Step 2:** Compute the right-tail probability:

$$
P(X > 5.1) = 1 - P(X \leq 5.1) = 1 - P(Z \leq 0.5)
$$

---

**Step 3 (Python):** You can calculate this using `norm.cdf()`:


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Parameters ---
mu = 5.0          # mean diameter
sigma = 0.2       # standard deviation
threshold = 5.1   # upper limit

# --- Step 1: Standardize ---
Z = (threshold - mu) / sigma
print(f"Standardized Z = {Z:.2f}")

# --- Step 2: Probabilities ---
P_left = norm.cdf(Z)
P_right = 1 - P_left

print(f"P(X <= {threshold}) = P(Z <= {Z:.2f}) = {P_left:.4f}")
print(f"P(X > {threshold}) = 1 - P(X <= {threshold}) = 1 - {P_left:.4f} = {P_right:.4f}")

# --- Step 3: Plot ---
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
pdf = norm.pdf(x, loc=mu, scale=sigma)

plt.figure(figsize=(8,5))

# PDF line
plt.plot(x, pdf, color='navy', lw=2, label=f'N({mu}, {sigma**2})')

# --- Shade succeed region (green) ---
x_fill_pass = np.linspace(mu - 4*sigma, threshold, 200)
plt.fill_between(x_fill_pass, 0, norm.pdf(x_fill_pass, mu, sigma),
                 color='green', alpha=0.5, label=f'Succeed (X â‰¤ {threshold})')

# --- Shade fail region (red) ---
x_fill_fail = np.linspace(threshold, mu + 4*sigma, 200)
plt.fill_between(x_fill_fail, 0, norm.pdf(x_fill_fail, mu, sigma),
                 color='red', alpha=0.5, label=f'Fail (X > {threshold})')

# Threshold line
plt.axvline(threshold, color='black', linestyle='--', lw=1.5)

# Labels, title, legend
plt.title('Right-Tail Probability: Rod Diameter Exceeding Threshold')
plt.xlabel('Rod Diameter (mm)')
plt.ylabel('Density f(x)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Example 4 â€” Particle Energy Measurement

A particle detector measures the kinetic energy \(X\) of a particle beam. Due to experimental 
noise, the measured energy follows:

$$
X \sim N(\mu=100\ \text{MeV}, \sigma^2 = 9)
$$

The detector has a reliable operating range: \(95 \le X \le 105\) MeV.  
**Question:** What is the probability that a particle's measured energy falls **outside** this range?

---

**Step 1:** This is **not a standard normal**.

**Step 2:** Standardize the bounds:

$$
Z_{\text{low}} = \frac{95 - 100}{3} = -1.67, \quad
Z_{\text{high}} = \frac{105 - 100}{3} = 1.67
$$

---

**Step 3:** Probabilities:

- Success (within range): $P(95 \le X \le 105) = P(Z_{\text{low}} \le Z \le Z_{\text{high}})$  
- Failure (outside range): $P(X < 95 \text{ or } X > 105)$ = $1 - P(95 \le X \le 105)$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Parameters ---
mu = 100.0       # mean energy (MeV)
sigma = 3.0      # standard deviation (sqrt of 9)
lower = 95.0     # lower bound
upper = 105.0    # upper bound

# --- Standardize bounds ---
Z_lower = (lower - mu) / sigma
Z_upper = (upper - mu) / sigma
print(f"Z_lower = {Z_lower:.2f}, Z_upper = {Z_upper:.2f}")

# --- Probabilities ---
P_pass = norm.cdf(Z_upper) - norm.cdf(Z_lower)
P_fail = 1 - P_pass
print(f"P(Energy outside range) = {P_fail:.4f}")
print(f"P(Energy within range) = {P_pass:.4f}")

# --- Plot PDF ---
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
pdf = norm.pdf(x, loc=mu, scale=sigma)

plt.figure(figsize=(8,5))

# PDF line
plt.plot(x, pdf, color='navy', lw=2, label=f'N({mu}, {sigma**2})')

# --- Shade succeed region (green) ---
x_fill_pass = np.linspace(lower, upper, 200)
plt.fill_between(x_fill_pass, 0, norm.pdf(x_fill_pass, mu, sigma),
                 color='green', alpha=0.5, label='Success (95 â‰¤ X â‰¤ 105)')

# --- Shade fail regions (red) ---
x_fill_fail_low = np.linspace(mu - 4*sigma, lower, 200)
plt.fill_between(x_fill_fail_low, 0, norm.pdf(x_fill_fail_low, mu, sigma),
                 color='red', alpha=0.5, label='Fail (X < 95)')

x_fill_fail_high = np.linspace(upper, mu + 4*sigma, 200)
plt.fill_between(x_fill_fail_high, 0, norm.pdf(x_fill_fail_high, mu, sigma),
                 color='red', alpha=0.5, label='Fail (X > 105)')

# Threshold lines
plt.axvline(lower, color='black', linestyle='--', lw=1.5)
plt.axvline(upper, color='black', linestyle='--', lw=1.5)

# Labels, title, legend
plt.title('Two-Tailed Probability: Particle Energy Measurement')
plt.xlabel('Measured Energy (MeV)')
plt.ylabel('Density f(x)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
